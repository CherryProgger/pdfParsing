```python
import fitz
from PIL import Image
import io
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTImage, LAParams
from pdfminer.pdfinterp import resolve1
import os
import logging
import multiprocessing
```

```python
def extract_raw_images(pdf_path, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    doc = fitz.open(pdf_path)
    for i in range(len(doc)):
        for img_index, img in enumerate(doc.get_page_images(i)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            image_stream = io.BytesIO(image_bytes)
            try:
                image = Image.open(image_stream)
                image_path = os.path.join(output_dir, f"page_{i+1}_image_{img_index+1}.{image_ext}")
                image.save(image_path)
                print(f"Изображение сохранено: {image_path}")
            except Exception as e:
                print(f"Ошибка при сохранении изображения на странице {i + 1}: {e}")
```

```python
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
```

```python
def extract_text_from_page(page):
    """Извлечение текста с страницы."""
    return page.get_text("text")
```

```python
def extract_images_from_page(page, page_number):
    """Извлечение изображений с страницы с помощью PyMuPDF."""
    images = []
    blocks = page.get_text("dict")["blocks"]
    for block in blocks:
        if block["type"] == 1 and "xref" in block:
            try:
                xref = block["xref"]
                base_image = page.extract_image(xref)
                if base_image:
                    image_bytes = base_image["image"]
                    image_stream = io.BytesIO(image_bytes)
                    image = Image.open(image_stream)
                    images.append(image)
            except Exception as e:
                logging.error(f"Ошибка при обработке изображения на странице {page_number + 1} (xref={xref}): {e}")
    return images

```

```python
def process_page(pdf_path, page_number):
    """Обработка страницы для извлечения текста и изображений."""
    logging.info(f"Обработка страницы {page_number + 1}")
    pdf_document = fitz.open(pdf_path)
    page = pdf_document.load_page(page_number)

    text = extract_text_from_page(page)
    images = extract_images_from_page(page, page_number)

    pdf_document.close()

    # Если PyMuPDF не нашёл изображения, используем pdfminer.six
    if not images:
        logging.info(f"PyMuPDF не нашёл изображения на странице {page_number + 1}, используем pdfminer.six")
        for page_layout in extract_pages(pdf_path, page_numbers=[page_number], laparams=LAParams()):
            for element in page_layout:
                if isinstance(element, LTImage):
                    try:
                        image_stream = element.stream
                        image_bytes = resolve1(image_stream).get_rawdata()
                        if image_bytes:
                            image_stream = io.BytesIO(image_bytes)
                            image = Image.open(image_stream)
                            images.append(image)
                    except Exception as e:
                        logging.error(f"Ошибка при обработке изображения с pdfminer на странице {page_number + 1}: {e}")

    return {
        "page_number": page_number,
        "text": text,
        "images": images
    }
```

```python
def extract_text_and_images(pdf_path):
    # Проверка и создание папки для изображений
    output_dir = "output_images"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    pdf_document = fitz.open(pdf_path)
    num_pages = pdf_document.page_count
    pdf_document.close()

    # Используем multiprocessing для параллельной обработки страниц
    with multiprocessing.Pool(processes=multiprocessing.cpu_count()) as pool:
        results = pool.starmap(process_page, [(pdf_path, i) for i in range(num_pages)])

    return results
```

```python
data = extract_text_and_images("/content/drive/MyDrive/Colab/инструкция_8DJH.pdf")
for page in data:
    print(f"Страница {page['page_number'] + 1}")
    print("Текст:", page['text'])
    print(f"Изображений: {len(page['images'])}")
    for i, image in enumerate(page['images']):
        try:
            image.save(f"/content/drive/MyDrive/Colab/output_images/page_{page['page_number'] + 1}_image_{i + 1}.jpg")
        except Exception as e:
            print(f"Не удалось сохранить изображение {i + 1} на странице {page['page_number'] + 1}: {e}")
```
